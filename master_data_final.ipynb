## 카카오 API로 좌표 -> 주소

```python
from dotenv import load_dotenv
import os, requests, time, pprint

load_dotenv()

KAKAO_API_KEY = os.getenv('KAKAO_API_KEY')
KAKAO_API_KEY = 'bd640091d4a1413f6c3789022f3e8f8a'
```

```python
addr_request_fail = 0
index_error = 0
data = {}

url = "https://dapi.kakao.com/v2/local/geo/coord2address.json"

headers = {
    'Authorization': f'KakaoAK {KAKAO_API_KEY}'
}


# GET 요청 보내기
for idx, stn_data in false_df.iterrows():
    params = {
        'input_coord': 'WGS84',
        'x': stn_data['stn_lng'],
        'y': stn_data['stn_lat']
    }

    
    response = requests.get(url, headers=headers, params = params)

    # 응답 확인
    if response.status_code == 200:
        data = response.json()
        try:
            
            new_addr = data['documents'][0]["address"]["address_name"]
            
            if("서울 " in new_addr):
                false_df.at[idx, "stn_addr"] = new_addr.replace("서울 ", "서울특별시 ")
            else:
                false_df.at[idx, "stn_addr"] = new_addr.replace("경기 ", "경기도 ")
            false_df.at[idx, "stn_gu"] = data['documents'][0]["address"]["region_2depth_name"]
        except:
            index_error += 1
            pprint.pprint(data)
            print()
    else:
        addr_request_fail += 1
        print(f"요청 실패: {response.status_code}")

print(f'addr_request_fail: {addr_request_fail}')
```
